Import faiss   and numpy libraries

```python
import numpy as np
import faiss
```

Using the rand() function from numpy, create a set of
100 000 numeric vectors of length 768 with random numbers as elements.   Store a set of numeric vectors in the 'data' variable.

```python
data = np.random.rand(100_000, 768)

print("Shape of data:", data.shape)
```

Using the simple, very accurate (but not too fast) 'IndexFlatL2()' method from the faiss library, create an index for vectors of length 768 and store its object in the variable 'index'.

```python
index = faiss.IndexFlatL2(768)
```

Using the 'assert' statement and the 'is_trained' property of the 'index' object to check if the index is created.

```python
assert index.is_trained
print(index.is_trained)
```

Index your vector set using the .add() method of the 'index' object, calling the method with the 'data' variable as an argument.

```python
index.add(data)

```

Using the 'index' property of the 'ntotal' object, check how many vectors are contained at the index.

```python
print("Number of vectors in index:", index.ntotal)
```

Using the rand() function from numpy, create a  numeric vectors of length 768 with random numbers as elements. Store a numeric vectors in the 'query' variable.

```python
query = np.random.rand(768)
```

Search the created index for similar vectors for the 'query' vector. Use the .search(query, n) method of the 'index' object with two arguments: the query vector and the number of similar vectors to search for.  
Store the search results in two variables: D (data) and I (indixes)  using multiple assignment.

```python
import numpy as np

# reshape
query_2d = query.reshape(1, -1)

# Search for similar vectors
D, I = index.search(query_2d, 5)
```

Print the results

```python
print("Distances (D):")
print(D)

print("\nIndexes (I):")
print(I)
```
